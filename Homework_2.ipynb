```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def multi_step_method(y0, T, dt, a):
    t_values = np.arange(0, T + dt, dt)
    y_values = np.zeros(len(t_values))

    y_values[0] = y0
    y_values[1] = np.e**(dt**2)

    for i in range(2, len(t_values)):
        fn = 2 * t_values[i-1] * y_values[i-1]
        fn_minus_1 = 2 * t_values[i-2] * y_values[i-2]

        y_values[i] = (1+a)*y_values[i-1] - a*y_values[i-2] + (dt/2)*((3-a)*fn - (1+a)*fn_minus_1)

    return y_values, t_values
```

C). $a=0$ and $\Delta t=0.1$ and $0\leq t\leq 2$

```python
y1, t1 = multi_step_method(1, 2, 0.1, 0)

plt.plot(t1, y1, label='Approximation at a=0')
plt.plot(t1, np.e**t1, label='Exact Value')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
```

D). $a=-5$ and $\Delta t=0.1$ and $0\leq t\leq 2$

```python
y1, t1 = multi_step_method(1, 2, 0.1, -5)

plt.plot(t1, y1, label='Approximation at a=0')
plt.plot(t1, np.e**t1, label='Exact Value')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
```

```python
y1, t1 = multi_step_method(1, 2, 0.01, -5)

plt.plot(t1, y1, label='Approximation at a=0')
plt.plot(t1, np.e**t1, label='Exact Value')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
```

We can see that the approximation does get better as $\Delta t$ is decreased, because the method is unstable we will never get a perfect approximation, but having a smaller $\Delta t$ makes this issue less prevelent because the local truncation error becomes smaller.
